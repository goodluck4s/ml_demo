# 探索变量间的关系
比统计更直观呢~

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
np.random.seed(sum(map(ord, "regression")))
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips[tips['size']==1]
```

## 绘制线性回归模型
最简单的方式：散点图 + 线性回归 + 95%置信区间

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

对于变量离线取值，散点图就显得有些尴尬了。。。

```python
sns.lmplot(x="size", y="tip", data=tips)
```

方法1：加个小的抖动

```python
sns.lmplot(x="size", y="tip", data=tips, x_jitter=.08)  #否则很多数堆叠在一起 看不出到底有多少个
```

方法2：离散取值上用均值和置信区间代替散点

```python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean)  #对tips求均值（点）和标准差（点上下的竖线）
```

## 拟合不同模型
有些时候线性拟合效果不错，有些时候差强人意~

```python
anscombe = sns.load_dataset("anscombe") 
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"), ci=None, scatter_kws={"s": 80})  #ci=None 置信区间不用画了
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"), ci=None, scatter_kws={"s": 80})
```

试试高阶拟合~

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"), order=2, ci=None, scatter_kws={"s": 80})  #order=2 用2阶拟合
```

异常值肿么办？

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"), robust=True, ci=None, scatter_kws={"s": 80}) 
#robust=True 鲁棒性保证，排除异常值的干扰 但是这个功能慎用，因为这个功能是函数里写死的
```

二值变量如何拟合？

```python
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x="total_bill", y="big_tip", data=tips, y_jitter=.05)  #不让他们都堆叠在0 1  #这里是线性回归来拟合结果
```

```python
sns.lmplot(x="total_bill", y="big_tip", data=tips, logistic=True, y_jitter=.03, ci=None) #这里是用逻辑回归来拟合结果
```

如何评价拟合效果？残差曲线~

```python
data=anscombe.query("dataset == 'II'")
data
```

```python
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'I'"), scatter_kws={"s": 80})
```

拟合的好，就是白噪声的分布$N(0,\sigma^2)$
拟合的差，就能看出一些模式

```python
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'II'"), scatter_kws={"s": 80})
```

## 变量间的条件关系摸索

```python
sns.lmplot(x="total_bill", y="tip",hue="day", data=tips)#线的斜率越高 越大方
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, markers=["o", "x"])
```

尝试增加更多的分类条件

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", data=tips)
```
