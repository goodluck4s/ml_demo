### 1.谈谈您对数据可视，以及其对数据分析，机器学习的意义的理解

数据可视是通过具有统计涵义的图形来展示数据的方法，数据可视最重要的意义是帮助人们更好地理解数据，从而从数据中获得信息与价值。数据可视化的图形有多种，例如：折线图、直方图、柱状图、散点图、饼图、K线图、箱线图等等。

##### 数据可视对数据分析的意义：
- 数据可视是数据分析的一种手段，旨在借助具有丰富内涵的图形去揭示数据背后真正的信息。
- 数据可视化可做多种分析，例如：
   * 可通过折线图进行趋势分析；
   * 利用直方图、箱线图观察数据的分布情况；
   * 可通过消维汇总展示并分析数据在各个维度上的分布情况；
   * 可分析两个特征之间的相关性；
   * 等等。

##### 数据可视对机器学习的意义：
- 数据可视对机器学习最重要的意义是通过数据可视化去理解样本数据，提取样本特征，以更好地训练数据。 
- 机器学习训练数据的结果可借助可视化来进行分析，从而返回进一步优化机器学习模型，以得到更优的训练结果。

### 3.	一组数据的分布特征可以从哪几个方面进行描述？

数据分布的特征可以从三个方面进行测度和描述： 
- 1.分布的集中趋势，反映各数据向其中心值靠拢或聚集的程度。例如：众数、中位数、分位数、均值、几何平均数、切尾均值。
- 2.分布的离散程度，反映各数据远离其中心值的趋势。例如：极差（最大值和最小值之差）、内距（上四分位数与下四分位数之差）、方差、标准差、离散系数。
- 3.分布的形状，反映数据分布的偏度和峰度。
    - 偏度是数据分布偏斜程度的测度，包括对称、左偏和右偏；
    - 峰度是数据分布扁平程度的测度，峰度系数等于3则偏度适中，大于3是峰尖分布，小于3是扁平分布。

### 1.航班乘客变化分析

- 载入相应绘图库及航班乘客数据（flights）

```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns

mpl.rc('figure',figsize=(12,9))
mpl.rc('font',size=15)
mpl.rc('axes.spines',top=False, right=False)
```

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

- 分析年度乘客总量变化情况（折线图）

```python
tmp = data[['year','passengers']].groupby('year').sum()
tmp.reset_index(inplace = True)
sns.set(style = "ticks")
sns.factorplot(x = "year", y = "passengers", data = tmp, size = 7, aspect = 2 )
```

```python
#或者用pointplot函数
fig, ax = plt.subplots(figsize = (14,5))
sns.pointplot(x = "year", y = "passengers", data = tmp, ax = ax)
```

#### 分析结论：年度乘客总量变化几乎是稳步增加的趋势，在1953年和1958年增加的较缓慢。

- 分析乘客在一年中各月份的分布（柱状图）

```python
sns.factorplot(x='month',y='passengers',data=data, kind='bar',size=8)
plt.xticks(rotation=60)
```

```python
#或者使用barplot方法
fig, ax = plt.subplots(figsize = (12,8))
sns.barplot(x = "month", y = "passengers", data = data, ax = ax)
```

#### 分析：
- 一年当中，7月和8月的乘客数量几乎相同，且在这两个月份的乘客数量最多，为高峰期；
- 3、4、5、10月的乘客数量相当。
- 11月份的乘客数量最少，且从8月到11月，乘客数量逐步减少，直到12月份，乘客数量增加。
- 虽然7、8月份乘客数量最多，但各年7、8月份的乘客数量波动较大，不是十分稳定

### 2.鸢尾花花型尺寸分析

* 载入iris数据集
* 萼片（sepal）和花瓣（petal）的大小关系（散点图）
- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

* 载入iris数据集

```python
data = sns.load_dataset("iris")
data.head()
# 萼片长度，萼片宽度，花瓣长度，花瓣宽度，种类
```

* 萼片（sepal）和花瓣（petal）的大小关系（散点图）

```python
data['sepal_area'] = data['sepal_length']*data['sepal_width']
data['petal_area'] = data['petal_length']*data['petal_width']
data.head()
```

```python
sns.set(style = 'darkgrid')
sns.jointplot(x = "sepal_area", y = "petal_area", data = data, size = 7, ratio = 7)
```

#### 分析：
- 萼片（sepal）和花瓣（petal）的大小关系
- 当萼片较小（面积小于2）时，萼片大小和花瓣大小几乎是无相关性；
- 当萼片较大（面积大于2）时，萼片大小和花瓣大小是正相关的，即萼片随着花瓣变大而变大。



不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）

```python
#将两个度量用两个维属性标识，即增加一个维度字段
tmp1 = data[['species','sepal_area']]
tmp1['area_type'] = 'sepal_area'
tmp1.columns = ['species','area','area_type']
tmp2 = data[['species','petal_area']]
tmp2['area_type'] = 'petal_area'
tmp2.columns = ['species','area','area_type']
tmp = tmp1.append(tmp2, ignore_index=True)
tmp.head()
```

```python
#柱状图
sns.set(style = "ticks")
fig, ax = plt.subplots(figsize = (11,5)) #设置图的大小
sns.barplot(x = "species", y = "area", hue = "area_type", data = tmp,ax = ax)
#barplot会直接按x汇总，对y求平均，且默认展示方差。
```

```python
#箱式图
sns.set_style('whitegrid')
sns.boxplot(x = "species", y = "area", hue = "area_type", data = tmp)
```

#### 分析结论：
总体上来看
- 每一种种类都是sepal的大小比petal大；
- 就petal来看，种类按照virginica、versicolor、setosa，petal依次减小；
- 就sepal来看，种类按照virginica、setosa、versicolor，sepall依次减小；

```python
pclass_alive = data[['pclass','alive']].groupby(['pclass','alive']).size().unstack()
pclass_alive['total'] = pclass_alive['yes']+pclass_alive['no']
pclass_alive.reset_index(inplace = True)
pclass_alive
```

```python
#seaborn中没有现成的堆叠柱状图方法
fig, ax = plt.subplots(figsize = (11,5)) #设置图的大小
sns.barplot(x = "pclass", y  = "total", data = pclass_alive, color = "r", ax = ax)
bottom_plot = sns.barplot(x = "pclass", y  = "yes", data = pclass_alive, color = "b", ax = ax)
alive_bar = plt.Rectangle((0,0),1,1,fc='b',  edgecolor = 'none')
not_alive_bar = plt.Rectangle((0,0),1,1,fc='r',  edgecolor = 'none')
plt.legend([alive_bar, not_alive_bar], ['alive', 'not alive'], ncol = 2, prop={'size':16},loc = 'upper center')
```
